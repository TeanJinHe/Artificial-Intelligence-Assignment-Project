2. Create simple codes for fuzzy numbers below and the defuzzification using centroid defuzzification
method only. Plot the fuzzy numbers with the defuzzification results.

```python
#a)

import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

x = np.arange(0, 10.05, 0.1)
mfx = fuzz.trimf(x, [4, 7, 8.6])


defuzz_centroid = fuzz.defuzz(x, mfx, 'centroid')


labels = ['centroid']
xvals = [defuzz_centroid]
colors = ['pink']
ymax = [fuzz.interp_membership(x, mfx, i) for i in xvals]


plt.figure(figsize=(8, 5))

plt.plot(x, mfx, 'k')
for xv, y, label, color in zip(xvals, ymax, labels, colors):
    plt.vlines(xv, 0, y, label=label, color=color)
plt.ylabel('Fuzzy membership')
plt.xlabel('Universe variable (arb)')
plt.ylim(-0.1, 1.1)
plt.legend(loc=2)

plt.show()
```

```python
#b)

x = np.arange(0, 10.05, 0.1)
mfx = fuzz.trapmf(x, [5.5, 5.7, 6, 6.3])


defuzz_centroid = fuzz.defuzz(x, mfx, 'centroid')


labels = ['centroid']
xvals = [defuzz_centroid]
colors = ['indigo']
ymax = [fuzz.interp_membership(x, mfx, i) for i in xvals]


plt.figure(figsize=(8, 5))

plt.plot(x, mfx, 'k')
for xv, y, label, color in zip(xvals, ymax, labels, colors):
    plt.vlines(xv, 0, y, label=label, color=color)
plt.ylabel('Fuzzy membership')
plt.xlabel('Universe variable (arb)')
plt.ylim(-0.1, 1.1)
plt.legend(loc=2)

plt.show()
```
